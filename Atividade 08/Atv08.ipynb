```python
from random import randrange, uniform
from mpi4py import MPI
import math
import time

def integ_ex_1 (x):
    return (4 / (1 + pow(x,2)))

def integ_ex_2 (x, y, z):
    return (pow(z,2) + math.pow((math.sqrt(pow(x,2) + pow(y,2)) - 3), 2))

def check_area_integ_2 (x, y, z):
    return ((pow(z,2) + math.pow((math.sqrt(pow(x,2) + pow(y,2)) - 3), 2)) <= 1.0)

def monte_carlo_integ_ex_1 (n):
    total = 0

    for i in range(0,n):
        x = uniform(0.0, 100.0) # aleatorio entre 0 e 100
        total = total + integ_ex_1(x)

    return total / n

def est_erro (f1, f2, n, v):
    return (v * math.sqrt((f1 - f2) / n))

def monte_carlo_integ_ex_2 (n):

    total = 0.0
    div = 0

    while (div != n):
        x = uniform(1.0, 4.0)  # aletorio entre 1.0 e 4.0
        y = uniform(-3.0, 4.0) # aleatorio entre -3.0 e 4.0
        z = uniform(-1.0, 1.0) # aleatorio entre -1.0 e 1.0

        if(check_area_integ_2(x,y,z)): 
            total = total + integ_ex_2(x, y, z)
            div = div + 1

    return total / div

if __name__ == '__main__':
    
    comm = MPI.COMM_WORLD  # estabelece comunicacao
    proc = comm.Get_size() # retorna o numero de processos da comunicacao
    rank = comm.Get_rank() # retorna o id do processo atual
    
    print("\proc: " + str(proc) + " rank: " + str(rank) + "\n")
    
    # variaveis do exercicio 1
    a = 0;
    b = 1;

    v = [10, 100, 1000, 10000, 20000, 30000]
    
    print("|-----> Metodo de Monte Carlo - Exercicio 1")
    print("\tV = " + str(b - a) + "\n")

    for i in range(len(v)):

        print("\tn = " + str(v[i]))
        
        start_time = time.time() 
        mc_ex1 = monte_carlo_integ_ex_1(v[i])
        end_time = time.time()
        
        print("\tMonte Carlo = " + str((b - a) * mc_ex1))
        print("\tErro = " + str(est_erro(mc_ex1, pow(mc_ex1, 2), v[i], (b - a))))
        print("\tTempo = " + str(end_time - start_time)  + "\n")

    # variaveis do exercicio 2
    V = (1 * 4) * (- 3 * 4) * (- 1 * 1)

    print("\n|-----> Metodo de Monte Carlo - Exercicio 2" )
    print("\tV = " + str(V) + "\n")

    for i in range(len(v)):
        print("\tn = " + str(v[i]))
        
        start_time = time.time()
        mc_ex2 = monte_carlo_integ_ex_2(v[i])
        end_time = time.time()
        
        print("\tMonte Carlo = " + str(V * mc_ex2))
        print("\tErro = " + str(est_erro(mc_ex2, pow(mc_ex2, 2), v[i], V)))
        print("\tTempo = " + str(end_time - start_time) + "\n")

    print("-------------------------- MPI --------------------------")
    print("|-----> Metodo de Monte Carlo - Exercicio 1")
    print("\tV = " + str(b - a) + "\n")

    for i in range(len(v)):

        print("\tn = " + str(v[i]))
        
        start_time = time.time() 
        mc_ex1 = comm.reduce(monte_carlo_integ_ex_1(v[i]), root = 0) # soma os valores de todos os processos
        end_time = time.time()
        
        print("\tMonte Carlo = " + str(((b - a) * (mc_ex1 / proc)))) # retorna a media obtida pelos processos
        print("\tErro = " + str(est_erro((mc_ex1/proc), pow((mc_ex1/proc), 2), v[i], (b - a))))
        print("\tTempo = " + str(end_time - start_time)  + "\n")

    # variaveis do exercicio 2
    V = (1 * 4) * (- 3 * 4) * (- 1 * 1)

    print("\n|-----> Metodo de Monte Carlo - Exercicio 2" )
    print("\tV = " + str(V) + "\n")

    for i in range(len(v)):
        print("\tn = " + str(v[i]))
        
        start_time = time.time()
        mc_ex2 = comm.reduce(monte_carlo_integ_ex_2(v[i]), root = 0) # soma os valores de todos os processos
        end_time = time.time()
        
        print("\tMonte Carlo = " + str((V * (mc_ex2/proc))))  # retorna a media obtida pelos processos
        print("\tErro = " + str(est_erro((mc_ex2/proc), pow((mc_ex2/proc), 2), v[i], V)))
        print("\tTempo = " + str(end_time - start_time) + "\n")
```
